```python
df = pd.read_csv(path1 + '/페트병_테스트.csv')      # 공유폴더의 페트병_1039 파일을 내 드라이브에 옮겨 읽어야함
df
```

```python
for j in list(df.columns[5:10]):
    print(j)
```

```python
for j in list(df.columns[5:10]):
    file_l = []
    ind_list=[]
    df1 = df[df[j]==1]    ## 원하는 조건의 데이터프레임
    df2 = df1.reset_index()       ## df1 과정에서 인덱스가 꼬였기 때문에 reset 시키고 기존 인덱스를 하나의 컬럼으로 만듦
    
    for i in range(len(df2['index'])):
        ind_list.append(df2['index'][i])
    
    for k in ind_list:
        name = path1 + '/페트_페트병/PET_PET.' + str(k) + '.jpg' ### 페트병 (오류) 가 있는 경로 인덱스값은 int값이므로 str으로 바꾸어야함
        a = glob(name)
    file_l.extend(a)
    
    for o in range(len(file_l)):
        x = plt.imread(file_l[o])
        cv2.imwrite(path2 + j + '/PET_' + j + '.' + str(o) + '.jpg', x)
        plt.imshow(x)    

#file_l
```
