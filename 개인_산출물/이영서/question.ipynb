```python
question_list = []
for page in range(1,11):
    print('page', page)
    url = f"{base_url}{quote('애매한 분리수거')}&page={page}"
    driver.get(url)
    time.sleep(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    lis = soup.select('ul.basic1 > li')

    for li in lis:
        sub_url = li.select_one('a._searchListTitleAnchor')['href']
        driver.get(sub_url)
        time.sleep(2)

        count = int(driver.find_element_by_css_selector('._answerCount.num').text)
        answer_page = (count - 1) // 5
        #print(count, end=', ')
        for i in range(answer_page):
            driver.find_element_by_id('nextPageButton').click()
            time.sleep(2)

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        text = driver.find_element(By.CLASS_NAME, 'c-heading__content').text
        question_list.append(text)

        driver.back()
        time.sleep(2)
```

```python
stopwords ="""
    환경 생각 마음 아주 제대로 분리수거 질문 답변 분 포함 다른 과태료 납부
    등 스티커 류 사이 음식점 것 역시 두께 약간 음 처리 오염 원인 하나요 걔
    서울 관악구 빌라 원룸 살 아시 것 답변 주시 안 만능 관통 기 플라스틱 돈 
    철 아무 기능 높이 넓이 정도 집 안 그냥 용 종량제 고정 분리 건 저 길 나
    아빠 보고 진짜 제 좌 시 막 자가 격리 학생 슬슬 요일 뭘 이건 분류 떄 알 
    대해 좀 어디 온 겉 재활용 표시 절대 쓰레기 자원 집단 회수 곳 장소 쇼 식 종이크기 상관
    쓰레기봉투 마트 편의점 함 또 더 쪽 관리 실 뭐 거 맡 자꾸 스트레스 딱지 반
    거 가끔 보시 말 물건 거의 화장실 물건 거지 고장 그 조금 넣기 의류수거함 캔 지난 초콜릿 덕지덕지 
    가요 의문 환경보호 취지 전혀 거기 생 이듭 니 거기 때문 머 일수 상황 일이 증거 다시 차단 거주 데 
    제거 수축 그동안 경우 운영 오늘 옆 앞 담배 통화 후 이번 언어 관심 사회 방 건물 포장 축구 비 무용 
    층 카톡 사람 먼저 일 보 작품 저번 절 자기 자리 걸 부 불 알바 번 업체 한번 일단 다가 완전 
    도둑 첫 다이소 죄목 절도죄 점유이탈물횡령죄 전 크게 얘기 처음 문제 군소리 점 그때 탐구 길이 교육 사본 
    저건 오빠 버럭 화 내면 구 어이 말좀 하라 계속 이나 도데 체 왜 요 평소 내공 다 해설 머리스타일 추상 
    순화 거나 대체 작업 꼭 잡고 문장 만 습 디 굳이 은어 속어 잔재 외래어 주 애 난 놈 야간 토욜날 
    엊그제 제자리 여기 베란다 감 토요일 쓰기 입장 시작 끝 얼굴 택시 내 패턴 화요일 메카 투명 명작
    닉 스크랩 추가 보고서 살인 범인 내용증명 메시지 잡곡 초대 고지서 만약 주택 월요일 학교 통
    재미나 편의 합성어 유행어 목적 를 선택 고유 자국 노력 욕 통용 물질 방법 손해 에반게리온 아침
    헌 이후 크기 얼마 처분 여러가지 임 번호 서울시 대한 고소 우리 설거지 응 형 처럼 프레임 접이식 대형
    단어 대체어 사전 의미 따위 종류 별 로 건가 따라서 가름 나눔 배출 가능 론 로만 짝 통해 흰색 판타지 작성 개
    도통 검색 지식인 님들 인터넷 게 정부 개인 해결 방안 내일 은 사절 때 여부 그림 짜증 자 대충 
    아파트 집앞 전봇대 지금 확인 수 저희 최근 신경 달라 전체 공지 정말 구석 벌금 토요일 사료 만화 이름 
    청소 등등 이면 이해 문의 여자 무시 그거 중 품목 고 위 리필 나머지 부분 이 해 화가 사이트 
    가족 뭔가 사고 나이 오지 가장 수질 코믹 계약 헌터 슬레이어즈 소스 보라 기준 달 사가 막상 배상 
    포기 가야 눈 모레 롤 기숙사 환경오염 원래 법 염 생활 키 루머 머리카락 뉘앙스 궁금점 지정 보통 호수 
    다재 활용 기도 구분 괜찬 재질 도움 정리 이사해 일반 취습 기회 싹 세 가지 이사 여 물 방치 및 
    자취 립 박아 집기 예 돌리지 못 이제 치킨 동생 준비 숙제 영어 도 달래 사실 시간 지불 감독 
    둥 연성 저격 글 잘못 조건 매장 미연시 대책 손님 관계 폐기물 너 의 누나 통장 낼 시급 포 학년 아예 전화 
    제품 남자 혹시 외 가연성 관계 누가 뚜껑 부탁 몇 인정 아이 바로 진행 계좌 바람 핀 제목 이구 취향 상태 빨래 사과 
    입자 줄 시장 카페 동안 남차친구 하렘 이야기 각 절차 네 위해 부모님 남자친구 연락 법적 갑자기 어제 
"""

stop_words = stopwords.split()
new_tokens = []
for i in range(len(tokens)):
    for word in tokens[i]:
        if word not in stop_words:
            new_tokens.append(word)

#[word for word in tokens if word not in stop_words]
token_count = Counter(new_tokens)
```

```python
import pandas as pd
df = pd.DataFrame(columns=['물품', '개수'])
df['물품']= token_count.keys()
df['개수']= token_count.values()
```

```python
df_count = df.sort_values(by='개수',ascending=False).head(500)
df_count.reset_index(drop=True)
```
